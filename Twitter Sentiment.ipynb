```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("0000000000002747_training_twitter_x_y_train.csv",delimiter=',')
```

```python
df.head()
```

```python
df.shape
```

```python
sentiment=np.array(df["airline_sentiment"])
```

```python
import nltk
from nltk.corpus import stopwords
import string
```

```python
stop=stopwords.words("english")+list(string.punctuation)
```

```python
stop.append('')
```

```python
docs=[]
for i in range(df.shape[0]):
    l=df["text"][i].split(" ")
    l2=[]
    for j in l:
        j1=j.lower()
        if j1 not in stop:
            l2.append(j)
    docs.append((l2,sentiment[i]))
```

```python
from nltk.stem import WordNetLemmatizer
lem=WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
from nltk import pos_tag
```

```python
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('R'):#adverb
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
def clean_review(words):
    o=[]
    for w in words:
            pos=pos_tag([w])
            new_word=lem.lemmatize(w,pos=get_simple_pos(pos[0][1]))#pos[0][1] as pos_tag gives a 
                                                                   #tuple inside an array
            o.append(new_word.lower())
    return o
```

```python
clean_docs=[]
for i in docs:
    new_r=clean_review(i[0])
    clean_docs.append((new_r,i[1]))
```

```python
clean_docs[0]
```

```python
de=pd.read_csv("0000000000002747_test_twitter_x_test.csv",delimiter=',')
```

```python
docs2=[]
for i in range(de.shape[0]):
    l=de["text"][i].split(" ")
    l2=[]
    for j in l:
        j1=j.lower()
        if j1 not in stop:
            l2.append(j)
    docs2.append(l2)
```

```python
clean_docs2=[]
for i in docs2:
    new_r=clean_review(i)
    clean_docs2.append(new_r)
```

```python
clean_docs2[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
x_train=[" ".join(document) for document,category in clean_docs]
x_test=[" ".join(document) for document in clean_docs2]
```

```python
cvc=CountVectorizer(max_features=2000)
x_train_features=cvc.fit_transform(x_train)
x_train_features.todense()
```

```python
cvc.get_feature_names()
```

```python
x_test_features=cvc.transform(x_test)
```

```python
x_test_features.todense()
```

```python
x_test_features.shape
```

```python
y_train=[category for document,category in clean_docs]
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(penalty='l1')
lr.fit(x_train_features,y_train)
```

```python
yp=lr.predict(x_test_features)
```

```python
nltk.FreqDist(yp)
```
